```python
import pickle
import numpy as np

path = "/srv/xw4e25/dysarthria_explanation/data/fbank_ep04.pickle"

with open(path, "rb") as f:
    data = pickle.load(f)

print("Top-level type:", type(data))

if isinstance(data, dict):
    print("Number of top-level keys:", len(data))
    print("Top-level keys:", list(data.keys())[:20])
    print("=" * 80)

    for key in data.keys():
        value = data[key]

        print(f"\nTop-level key: {key}")
        print("Value type:", type(value))

        if hasattr(value, "shape"):
            print("Value shape:", value.shape)
            if hasattr(value, "dtype"):
                print("Value dtype:", value.dtype)

        elif isinstance(value, dict):
            print("Number of utterances:", len(value))
            utt_keys = list(value.keys())
            print("First 10 utterance keys:", utt_keys[:10])

            print("-" * 80)

            for utt in utt_keys[:10]:
                utt_value = value[utt]

                print(f"Utterance: {utt}")
                print("  Type:", type(utt_value))

                if hasattr(utt_value, "shape"):
                    print("  Shape:", utt_value.shape)
                    if hasattr(utt_value, "dtype"):
                        print("  Dtype:", utt_value.dtype)

                elif isinstance(utt_value, list):
                    print("  List length:", len(utt_value))
                    if len(utt_value) > 0:
                        print("  First item type:", type(utt_value[0]))
                        if hasattr(utt_value[0], "shape"):
                            print("  First item shape:", utt_value[0].shape)

                elif isinstance(utt_value, tuple):
                    print("  Tuple length:", len(utt_value))
                    for i, item in enumerate(utt_value):
                        print(f"    Item {i} type:", type(item))
                        if hasattr(item, "shape"):
                            print(f"    Item {i} shape:", item.shape)
                            if hasattr(item, "dtype"):
                                print(f"    Item {i} dtype:", item.dtype)

                else:
                    print("  Value:", utt_value)

                print()

        elif isinstance(value, list):
            print("List length:", len(value))
            if len(value) > 0:
                print("First item type:", type(value[0]))
                if hasattr(value[0], "shape"):
                    print("First item shape:", value[0].shape)

        else:
            print("Value:", value)

else:
    print("The loaded pickle is not a dict.")
    if hasattr(data, "shape"):
        print("Shape:", data.shape)
        print("Dtype:", data.dtype)
```

```python
import h5py

path = "/srv/xw4e25/dysarthria_explanation/data/TORGO_fbanks_data.h5"

with h5py.File(path, "r") as f:
    print("All keys in h5:")
    print(list(f.keys()))
    print("=" * 80)

    for key in f.keys():
        obj = f[key]

        print(f"\nKey: {key}")
        print("Type:", type(obj))

        if isinstance(obj, h5py.Dataset):
            print("Shape:", obj.shape)
            print("Dtype:", obj.dtype)

            first_item = obj[0]

            print("First item type:", type(first_item))

            if hasattr(first_item, "shape"):
                print("First item shape:", first_item.shape)
                print("First item dtype:", first_item.dtype)

            print("First item content:")
            print(first_item)

        elif isinstance(obj, h5py.Group):
            print("This is a group. Sub-keys:")
            print(list(obj.keys()))

        print("-" * 80)
```
